```python
from smolagents import tool

@tool
def solve_sudoku(grid: str) -> str:
    """
    Solves a 9x9 Sudoku puzzle and returns it in a nice ASCII-art format.

    Args:
        grid: A string of 81 digits representing the Sudoku board.
              Use '0' for empty cells.

    Returns:
        A string showing the solved Sudoku in a 9x9 ASCII grid.
    """

    # Convert input string to 2D board
    board = [[int(grid[r*9 + c]) for c in range(9)] for r in range(9)]

    # Function to check if placing n at (r, c) is valid
    def valid(r, c, n):
        if n in board[r]: return False
        if n in [board[i][c] for i in range(9)]: return False
        br, bc = (r//3)*3, (c//3)*3
        for i in range(3):
            for j in range(3):
                if board[br+i][bc+j] == n:
                    return False
        return True

    # Backtracking solver
    def solve():
        for r in range(9):
            for c in range(9):
                if board[r][c] == 0:
                    for n in range(1, 10):
                        if valid(r, c, n):
                            board[r][c] = n
                            if solve():
                                return True
                            board[r][c] = 0
                    return False
        return True

    solve()

    # ASCII-art formatter
    def format_board(board):
        lines = []
        for r in range(9):
            row = " ".join(str(board[r][c]) for c in range(9))
            # Add vertical separators for boxes
            row = row[:6] + "| " + row[6:12] + "| " + row[12:]
            lines.append("| " + row + " |")
            # Add horizontal separator after every 3 rows
            if r == 2 or r == 5:
                lines.append("+-------+-------+-------+")
        # Add top and bottom borders
        return "+-------+-------+-------+\n" + "\n".join(lines) + "\n+-------+-------+-------+"

    return format_board(board)
```

```python
from smolagents import CodeAgent, InferenceClientModel, tool
```

```python
agent = CodeAgent(
    tools=[solve_sudoku],
    model=InferenceClientModel()
)
```

```python
agent.run("Solve this sudoku: 060350000820000009105000308002009005080003097000006004070000002000014070006200000")
```
